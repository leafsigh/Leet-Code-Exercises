We are playing the Guess Game. The game is as follows:

<br>I pick a number from 1 to n. You have to guess which number I picked.

<br>Every time you guess wrong, I'll tell you whether the number I picked is higher or lower.

<br>However, when you guess a particular number x, and you guess wrong, you pay $x$. You win the game when you guess the number I picked.

### Example
n = 10, I pick 8.

<br>First round:  You guess 5, I tell you that it's higher. You pay $5$.
<br>Second round: You guess 7, I tell you that it's higher. You pay $7$.
<br>Third round:  You guess 9, I tell you that it's lower. You pay $9$.

<br>Game over. 8 is the number I picked.

<br>You end up paying $5 + $7 + $9 = $21.
Given a particular n â‰¥ 1, find out how much money you need to have to guarantee a win.

```python
# Time limit exceeded
memo={}
def getMoneyAmount(n):
    def helper(l,u):
        key=(l,u)
        print('key: ',key)
        if key in memo:
            print('memo: ',memo)
            return memo[key]
        if l>=u:
            print('l>=u, return: ',0)
            return 0
        ans=float("inf")
        for t in range(l,u+1):
            print('recur: t=',t,'l=',l,'u=',u)
            ans=min(ans,max(helper(t+1,u),helper(l,t-1))+t)
            print('ans: ',ans)
        memo[key]=ans
        print('memo: ',memo)
        return ans
    return helper(1,n)
```

```python
# bottom - up dynamic programming
def getMoneyAmount_2(n):
    need = [[0] * (n+1) for _ in range(n+1)]
    for lo in range(n, 0, -1):
        for hi in range(lo+1, n+1):
            print('low: ',lo,'high: ',hi)
            need[lo][hi] = min(x + max(need[lo][x-1], need[x+1][hi])
                               for x in range(lo, hi))
            print(need)
    return need[1][n]
```
